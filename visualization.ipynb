```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tqdm
from datetime import datetime
pd.set_option('display.max_rows', None)
```

```python
train = pd.read_parquet('data/trainprep.parquet')
test = pd.read_parquet('data/testprep.parquet')
val = pd.read_parquet('data/valprep.parquet')
```

```python
train['Type'].value_counts()
```

```python
dataframes = {'train': train, 'test': test, 'val': val}

for name, df in dataframes.items():
    if 'Type' in df.columns:
        type_counts = df['Type'].value_counts()

        plt.figure(figsize=(6, 6))
        colors = sns.color_palette('pastel')[0:len(type_counts)]  # Get colors from seaborn
        plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', colors=colors, startangle=90, wedgeprops={'edgecolor': 'black'})
        
        # Create a donut chart effect
        center_circle = plt.Circle((0, 0), 0.5, color='white')
        plt.gca().add_artist(center_circle)

        plt.title(f'Type Distribution in {name}')
        plt.show()
```

```python
numeric_cols = train.select_dtypes(include=[np.number]).columns
```

```python
# Histogram for each column
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(val[col], kde=True, bins=10, color='blue')
    plt.title(f'Histogram of {col} in val')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
val[numeric_cols].columns
```

```python
corr = ['Attack code_x', 'Avg source IP count', 'Type',
       'Detect count_y', 'Victim IP_y', 'Port number_y',  'Source IP count',
       'Packet speed_y_normalized', 'Data speed_y_normalized',
       'Avg packet len_y_normalized']
```

```python
correlation_matrix = val[corr].corr()

# Plot heatmap
plt.figure(figsize=(20, 15))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
# 📊 Boxplot: Show how numerical features vary by 'Type'
for col in val[numeric_cols]:
    if col not in ['Type', 'Type_encoded']:  # Skip non-numeric columns
        plt.figure(figsize=(10,10))
        sns.boxplot(x=val['Type'], y=val[col], palette='pastel')
        plt.title(f'{col} Distribution by Type')
        plt.show()
```
